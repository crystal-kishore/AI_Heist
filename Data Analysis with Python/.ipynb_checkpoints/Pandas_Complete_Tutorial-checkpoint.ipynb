```python
#Creating Own Function with Own logic

def cate_ath(row):
    if row['height_cm']<175 and row['weight_kg']<70:
        return 'Light Weight'
    elif row['height_cm']< 185 and row ['weight_kg']<=80:
        return 'Middle Weight'
    else:
        return 'Heavy Weight'
    
sports['Category']=sports.apply(cate_ath, axis=1)
```
